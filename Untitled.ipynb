```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bank-full.csv', sep = ';')
```

```python
df
```

```python
df.info()
```

```python
df.drop(columns=['contact', 'marital'], inplace = True)
```

```python
# job, marital, education, default, housing, loan, month, poutcome

from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder()

df['job'] = label_encoder.fit_transform(df['job'])
df['education'] = label_encoder.fit_transform(df['education'])
df['default'] = label_encoder.fit_transform(df['default'])
df['housing'] = label_encoder.fit_transform(df['housing'])
df['month'] = label_encoder.fit_transform(df['housing'])
df['loan'] = label_encoder.fit_transform(df['loan'])
df['poutcome'] = label_encoder.fit_transform(df['poutcome'])
df['y'] = label_encoder.fit_transform(df['y'])

df
```

```python
df.corr()
```

```python
df.drop(columns=['age', 'education', 'job', 'default', 'day'], inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Preprunning

parameters = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1, 2, 3, 4, 5],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
cv = GridSearchCV(tree_model, param_grid = parameters, cv = 5, scoring = 'accuracy')

cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_pred, y_test)
```
